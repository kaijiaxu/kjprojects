```
#Connect Four is a two-player connection board game in which the players first choose a color and then take turns dropping one colored disc from the top into a seven-column, six-row vertically suspended grid. The pieces fall straight down, occupying the lowest available space within the column. The objective of the game is to be the first to form a horizontal, vertical, or diagonal line of four of one's own discs. 

#Welcome message
print("Welcome to Connect4! The first to get 4 in a row to wins.")

def checkFourInRow(board):
    boardHeight = len(board[0])
    boardWidth = len(board)
    # check horizontal spaces
    for y in range(boardHeight):
        for x in range(boardWidth - 3):
            if board[x][y] == board[x+1][y] == board[x+2][y] == board[x+3][y] == 'R' != '.':
                return True
            elif board[x][y] == board[x+1][y] == board[x+2][y] == board[x+3][y] == 'B' != '.':
                return True
    # check vertical spaces
    for x in range(boardWidth):
        for y in range(boardHeight - 3):
            if board[x][y] == board[x][y+1] == board[x][y+2] == board[x][y+3] == 'R' != '.':
                return True
            elif board[x][y] == board[x][y+1] == board[x][y+2] == board[x][y+3] == 'B' != '.':
                return True

    # check / diagonal spaces
    for x in range(boardWidth - 3):
        for y in range(3, boardHeight):
            if board[x][y] == board[x+1][y-1] == board[x+2][y-2] == board[x+3][y-3] == 'R'!= '.':
                return True
            elif board[x][y] == board[x+1][y-1] == board[x+2][y-2] == board[x+3][y-3] == 'B' != '.':
                return True

    # check \ diagonal spaces
    for x in range(boardWidth - 3):
        for y in range(boardHeight - 3):
            if board[x][y] == board[x+1][y+1] == board[x+2][y+2] == board[x+3][y+3] == 'R' != '.':
                return True
            elif board[x][y] == board[x+1][y+1] == board[x+2][y+2] == board[x+3][y+3] == 'R'!= '.':
                return True

    return False

# Declaring rows 
N = 6
  
# Declaring columns 
M = 7
  
# using list comprehension  
# to initializing matrix 
board = [['.' for x in range(M)] for x in range(N)]
print("Current Game: ")
print(*(' '.join(row) for row in board), sep='\n')
print("1 2 3 4 5 6 7")

#loop through matrix while no 4-in-row
fourInRow = False

current_round = 1

while not fourInRow:
  #inital highest row (row not filled) = last row (at index 6)
  highestRow = 5
  #ask user to choose column
  if current_round%2 == 1:
    print("It's RED's turn!")
  else:
    print("It's BLUE's turn!")
  choice = int(input("Please choose the column which you want to place your piece 1-7\n"))
  #subtract 1 since indexes start from 0
  choice = choice - 1


#loop through matrix, get the highest row in column currently filled
  for i in range(N):
      if board[i][choice] != '.':
        #this box has already been chosen
        highestRow = i - 1
        break
  if current_round%2 == 1:
    board[highestRow][choice] = 'R'
    current_round+=1
  else:
    board[highestRow][choice] = 'B'
    current_round+=1

  #after update matrix, check if there is 4 in row
  fourInRow = checkFourInRow(board)
  print("Current Game: ")
  print(*(' '.join(row) for row in board), sep='\n')
  print("1 2 3 4 5 6 7")

print("Congratulations! You won! You managed to create 4 in a row.")
```
