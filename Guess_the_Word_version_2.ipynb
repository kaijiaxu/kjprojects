```
#Method 2 -- using arrays

i=0

while True:
  print("Welcome to Hangman! Guess the word in 10 rounds!")

  print("Word: _ _ _ _ _ _ ")

  x = ['purple', 'winter', 'orange', 'turtle', 'apples', 'people', 'school', 'commit', 'Africa', 'places']

  guess = ['_ ' for y in range(6)]

  current_round = 1
  
  #using arrays
  while True:
    word = x[i]

    character = input('Choose a letter\n')
    
    for j in range(6):
      if character == word[j-1]:
        guess[j-1] = word[j-1]

    print("Word: ", *(''.join(row) for row in guess), sep='')

    current_round += 1
  
    #convert guess array into string  
    str1 = ''
    for ele in guess:  
        str1 += ele 
    
    if word == str1:
      i+=1
      print("Congratulations! You found the word!")
      break
    
    elif current_round > 10:
      print("You lose :( Try to find the word within 10 tries next time.")
      break
```
